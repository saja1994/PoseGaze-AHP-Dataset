```python
@torch.no_grad()
def synthesize_novel_poses_and_expressions(expr=torch.zeros(100, dtype=torch.float32),
                                         pose=torch.zeros(15, dtype=torch.float32),
                                         image_size=(1024, 1024)):
    # Use CUDA if available
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"Using device: {device} for synthesis")

    # Get image dimensions
    img_h, img_w = image_size
    track_h, track_w = tr['image_size']

    # Calculate scaling factors
    fx_scale = max(track_h, track_w) * img_w / track_w
    fy_scale = max(track_h, track_w) * img_h / track_h
    cx_scale = img_w
    cy_scale = img_h

    # Create the intrinsics matrix on GPU with explicit float32
    cam_intrinsics = create_intrinsics_matrix(
        fx=tr["K"][0] * fx_scale,
        fy=tr["K"][0] * fy_scale,
        px=tr["K"][1] * cx_scale,
        py=tr["K"][2] * cy_scale,
    ).to(device).to(torch.float32)

    # Get neutral joint rotations and move to GPU
    rest_joint_rots = avatar._flame.get_neutral_joint_rotations()
    default_pose = torch.cat((
        rest_joint_rots["global"],
        rest_joint_rots["neck"],
        rest_joint_rots["jaw"],
        rest_joint_rots["eyes"],
        rest_joint_rots["eyes"]
    ), dim=0).to(device).to(torch.float32)

    # Move expr and pose to GPU with explicit float32
    expr = expr.to(device).to(torch.float32)
    pose = pose.to(device).to(torch.float32)

    # Create batch with all tensors on GPU with explicit float32
    batch = {
        'flame_shape': torch.from_numpy(tr["shape"][None]).to(device).to(torch.float32),
        'flame_expr': expr[None],
        'flame_pose': (pose + default_pose)[None],
        'flame_trans': torch.from_numpy(tr["translation"][[0]]).to(device).to(torch.float32),
        'cam_intrinsic': cam_intrinsics[None],
        'cam_extrinsic': torch.from_numpy(tr["RT"]).to(device).to(torch.float32)[None],
        'rgb': torch.zeros(1, 3, img_h, img_w, dtype=torch.float32).to(device)
    }

    # Ensure avatar and all its submodules are on GPU
    avatar.to(device)

    # Patch torch.arange globally
    original_arange = torch.arange

    def new_arange(*args, **kwargs):
        # Call the original arange and move result to device
        if 'device' not in kwargs:
            kwargs['device'] = device
        result = original_arange(*args, **kwargs)
        return result

    # Store patches to restore later
    patches = {
        'arange': (torch.arange, new_arange),
    }

    try:
        # Apply patches
        for name, (old_fn, new_fn) in patches.items():
            setattr(torch, name, new_fn)

        # Make prediction
        rgba = avatar.forward(batch, symmetric_rgb_range=False)
        shaded_mesh = avatar.predict_shaded_mesh(batch)

        return rgba, shaded_mesh

    finally:
        # Restore original functions
        for name, (old_fn, _) in patches.items():
            setattr(torch, name, old_fn)
```

```python
#Patient View

# Check if CUDA (GPU) is available and use it
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Head Movement Functions
def set_head_tilt(pose, degree, direction):
    """
    Tilt the head to the left or right by a specified degree.
    direction should be 'left' or 'right'.
    """
    degree =np.deg2rad(degree)
    if direction == 'right':
        pose[5] = degree
    elif direction == 'left':
        pose[5] = -degree
    else:
        raise ValueError("Direction must be 'left' or 'right'")
    return pose


def set_head_turn(pose, degree, direction):
    """
    Turn the head to the left or right by a specified degree.
    direction should be 'left' or 'right'.
    """
    degree =np.radians(degree)
    if direction == 'right':
        pose[4] = -degree
    elif direction == 'left':
        pose[4] = degree
    else:
        raise ValueError("Direction must be 'left' or 'right'")
    return pose

def set_chin_up(pose, expr, degree):
    """
    Move the chin up by a specified degree with optimized eye gaze and eyelid control.

    Args:
        pose: Pose tensor
        expr: Expression tensor to modify
        degree: Angle in degrees (positive value)

    Returns:
        Tuple of (pose, expr) with chin up and eye/eyelid compensation
    """
    # Convert degrees to radians (negative for chin up)
    rad_angle = -np.radians(degree)  # Negative for chin up

    # Set chin up (negative X-axis rotation)
    pose[3] = rad_angle

    # Calculate eye compensation - match the test sample exactly
    eye_comp_amount = 0.5  # Fixed value from test code
    pose[12] = eye_comp_amount  # Left eye down (positive)
    pose[9] = eye_comp_amount   # Right eye down (positive)

    # Apply exact eyelid parameters from the working test
    expr[8] = 4.0 # Eyelid parameter 1
    expr[9] = 4.0  # Eyelid parameter 2
    expr[10] = 1.0  # Supporting parameter 1
    expr[11] = 1.0  # Supporting parameter 2

    return pose, expr


def set_chin_down(pose, expr, degree):
    """
    Move the chin down by a specified degree with optimized eye gaze and eyelid control.

    Args:
        pose: Pose tensor
        expr: Expression tensor to modify
        degree: Angle in degrees (positive value)

    Returns:
        Tuple of (pose, expr) with chin down and eye/eyelid compensation
    """
    # Convert degrees to radians
    rad_angle = np.radians(degree)

    # Set chin down (positive X-axis rotation)
    pose[3] = rad_angle

    # Calculate eye compensation (using values from the test sample)
    eye_comp_amount = 0.27 * (degree / 12.0)  # Scaled value from test
    pose[12] = -eye_comp_amount  # Left eye up (negative)
    pose[9] = -eye_comp_amount   # Right eye up (negative)

    # Apply exact eyelid parameters from test
    expr[2] = -0.8 * (degree / 12.0)  # Left lower eyelid - raise up (negative)
    expr[3] = -0.8 * (degree / 12.0)  # Right lower eyelid - raise up (negative)
    expr[8] = -2.0 * (degree / 12.0)  # Negative value as found in testing
    expr[9] = -2.0 * (degree / 12.0)  # Negative value as found in testing

    return pose, expr

def pd_to_degrees(pd_value):
    """Convert prism diopters to degrees using the formula: degrees = arctan(PD/100)."""
    return np.degrees(np.arctan(pd_value/100))


## Patient View

def set_hypotropia(pose, pd_value, eye):
    """Simulate hypotropia (one eye lower than the other)."""
    # Convert from prism diopters to degrees, then to radians
    degree = np.radians(pd_to_degrees(pd_value))

    # Swap left and right for patient view
    if eye == 'right':  # Patient's right eye corresponds to observer's left
        pose[12] = degree  # Left eye down (positive)
        pose[9] = 0  # Keep right eye neutral
    elif eye == 'left':  # Patient's left eye corresponds to observer's right
        pose[12] = 0  # Keep left eye neutral
        pose[9] = degree  # Right eye down (positive)
    elif eye == 'bilateral':
        pose[12] = degree  # Both eyes down
        pose[9] = degree
    else:
        raise ValueError("Eye must be 'left', 'right', or 'bilateral'")

    # Zero out horizontal movements
    pose[13] = 0  # Left eye horizontal
    pose[10] = 0  # Right eye horizontal
    return pose

def set_hypertropia(pose, pd_value, eye):
    """Simulate hypertropia (one eye higher than the other)."""
    # Convert from prism diopters to degrees, then to radians
    degree = np.radians(pd_to_degrees(pd_value))

    # Swap left and right for patient view
    if eye == 'right':  # Patient's right eye corresponds to observer's left
        pose[12] = -degree  # Left eye up (negative)
        pose[9] = 0  # Keep right eye neutral
    elif eye == 'left':  # Patient's left eye corresponds to observer's right
        pose[12] = 0  # Keep left eye neutral
        pose[9] = -degree  # Right eye up (negative)
    elif eye == 'bilateral':
        pose[12] = -degree  # Both eyes up
        pose[9] = -degree
    else:
        raise ValueError("Eye must be 'left', 'right', or 'bilateral'")

    # Zero out horizontal movements
    pose[13] = 0  # Left eye horizontal
    pose[10] = 0  # Right eye horizontal
    return pose

def set_esotropia(pose, pd_value, eye):
    """Simulate esotropia (inward deviation of one or both eyes)."""
    # Convert from prism diopters to degrees, then to radians
    degree = np.radians(pd_to_degrees(pd_value))

    # Swap left and right for patient view
    if eye == 'right':  # Patient's right eye corresponds to observer's left
        pose[13] = degree  # Left eye inward (positive)
        pose[10] = 0  # Keep right eye neutral
    elif eye == 'left':  # Patient's left eye corresponds to observer's right
        pose[13] = 0  # Keep left eye neutral
        pose[10] = -degree  # Right eye inward (negative)
    elif eye == 'bilateral':
        pose[13] = degree  # Both eyes inward
        pose[10] = -degree
    else:
        raise ValueError("Eye must be 'left', 'right', or 'bilateral'")

    # Zero out vertical movements
    pose[12] = 0  # Left eye vertical
    pose[9] = 0  # Right eye vertical
    return pose

def set_exotropia(pose, pd_value, eye):
    """Simulate exotropia (outward deviation of one or both eyes)."""
    # Convert from prism diopters to degrees, then to radians
    degree = np.radians(pd_to_degrees(pd_value))

    # Swap left and right for patient view
    if eye == 'right':  # Patient's right eye corresponds to observer's left
        pose[13] = -degree  # Left eye outward (negative)
        pose[10] = 0  # Keep right eye neutral
    elif eye == 'left':  # Patient's left eye corresponds to observer's right
        pose[13] = 0  # Keep left eye neutral
        pose[10] = degree  # Right eye outward (positive)
    elif eye == 'bilateral':
        pose[13] = -degree  # Both eyes outward
        pose[10] = degree
    else:
        raise ValueError("Eye must be 'left', 'right', or 'bilateral'")

    # Zero out vertical movements
    pose[12] = 0  # Left eye vertical
    pose[9] = 0  # Right eye vertical
    return pose

```

```python
import os
import matplotlib.pyplot as plt
import json
import math

def save_results(row, views_results, base_directory, index, suffix=""):
    """
    Organized approach to save views and combined results with view names labeled in the combined view.
    Structure:
    base_directory/
        Diagnose/
            AHPType/
                case_{case_id}{suffix}/
                    view_name.png
                    combined_views.png
                    metadata.txt
    Args:
        row: DataFrame row with case information
        views_results: Generated view results
        base_directory: Base directory for saving outputs
        index: Current index in processing

        suffix: Optional suffix to add to directory name (default: "")
    """
    # Get case ID from the row
    case_id = row['CaseID']

    # Add suffix to case_id directory
    case_id_with_suffix = f"{case_id}{suffix}"

    # Create directory structure
    base_case_dir = os.path.join(base_directory, row['Diagnose'], row['AHPType'], case_id_with_suffix)
    os.makedirs(base_case_dir, exist_ok=True)

    # Save metadata
    metadata = {
        'case_id': case_id,
        'diagnosis': {
            'value': row['Diagnose'],
            'imputed': 'T' if row.get('PatientDiagnose_Imputed', '') == 'T' else 'F'
        },
        'ahp': {
            'type': {
                'value': row['AHPType'],
                'imputed': 'T' if row.get('AHPType_Imputed', '') == 'T' else 'F'
            },
            'direction': {
                'value': row['AHPDirection'],
                'imputed': 'T' if row.get('AHPDirection_Imputed', '') == 'T' else 'F'
            },
            'degree': {
                'value': row['AHPDegree'],
                'imputed': 'T' if row.get('AHPDegree_Imputed', '') == 'T' else 'F'
            }
        },
        'eye': {
            'side': {
                'value': row['Eye'],
                'imputed': 'T' if row.get('Eye_Imputed', '') == 'T' else 'F'
            },
            'misalignment': {
                'value': row['EyeMisalignment'],
                'imputed': 'T' if row.get('EyeMisalignment_Imputed', '') == 'T' else 'F'
            },
            'deviation': {
                'value': row['DegreePD'],
                'imputed': 'T' if row.get('DegreePD_Imputed', '') == 'T' else 'F'
            }
        },
        'processing_type': (
            'normal_head_with_eye_misalignment' if suffix == '_normal_head' else
            'head_pose_with_normal_eyes' if suffix == '_normal_eyes' else
            'combined'
        )
    }

    with open(os.path.join(base_case_dir, 'metadata.txt'), 'w') as f:
        json.dump(metadata, f, indent=4)

    # Save individual views (no labels)
    for view_name, rgba in views_results:
        fig, ax = plt.subplots(figsize=(10, 10))
        ax.imshow(rgba[0, :3].cpu().permute(1, 2, 0))
        ax.axis("off")
        view_filename = f"{view_name}.png"
        plt.savefig(os.path.join(base_case_dir, view_filename), bbox_inches='tight', pad_inches=0)
        plt.close()

    # Save combined view with enhanced labels
    num_views = len(views_results)
    num_cols = 4
    num_rows = math.ceil(num_views / num_cols)

    # Create figure with more space for captions
    fig, axes = plt.subplots(num_rows, num_cols, figsize=(30, 15))
    axes = axes.flatten()

    for i, (view_name, rgba) in enumerate(views_results):
        if i < len(axes):
            axes[i].imshow(rgba[0, :3].cpu().permute(1, 2, 0))
            axes[i].axis("off")

            # Enhanced caption styling - more prominent with better visibility
            axes[i].set_title(
                view_name,
                fontsize=16,       # Slightly larger font
                fontweight='bold', # Bold text
                pad=12,            # More padding
                color='white',
                backgroundcolor='black',
                bbox=dict(        # Add a box around the title
                    boxstyle='round,pad=0.5',
                    facecolor='black',
                    alpha=0.8,
                    edgecolor='white',
                    linewidth=1
                )
            )

    # Remove any extra axes if the grid is not completely filled
    for j in range(num_views, len(axes)):
        fig.delaxes(axes[j])

    # Add more space between subplots for better caption visibility
    plt.tight_layout(pad=3.0)
    plt.savefig(os.path.join(base_case_dir, f'combined_views{suffix}.png'), bbox_inches='tight', pad_inches=0.2)
    plt.close()

    return base_case_dir

def set_normal_head(pose):
    """
    Set the head to a completely normal position.
    Sets all head pose parameters to their correct default values.

    Args:
        pose: The pose tensor to modify

    Returns:
        The modified pose tensor with head in normal position
    """
    # Global rotation
    pose[0] = 0       # rot0
    pose[1] = np.pi   # rot1 - this is critical for front view
    pose[2] = 0       # rot2

    # Neck rotation
    pose[3] = 0       # neck0 - chin up/down
    pose[4] = 0       # neck1 - head turn left/right
    pose[5] = 0       # neck2 - head tilt left/right

    # Jaw movement
    pose[6] = 0       # jaw

    return pose


# Helper function for resetting head to neutral position
def reset_head_to_neutral(pose):
    """
    Reset the head to a completely neutral position.
    This ensures no head tilt, turn, or chin movement.
    """
    # Reset neck rotations to zero (straight head position)
    pose[3] = 0  # Neck X-axis rotation (chin up/down)
    pose[4] = 0  # Neck Y-axis rotation (head turn left/right)
    pose[5] = 0  # Neck Z-axis rotation (head tilt left/right)
    return pose
```

```python
##observer view

def process_eye_misalignment_only(data_file, device, base_directory, start_index=0):
    """
    Process eye misalignment with normal head pose (no head movement).

    Args:
        data_file: Path to the CSV file
        device: PyTorch device to use
        base_directory: Base directory for saving outputs
        start_index: Index to start processing from (default: 0)
    """
    # Load and process data
    df = pd.read_csv(data_file)

    # Remove all rows where 'Eye' is 'Unilateral'
    df = df[df['Eye'].str.strip().str.lower() != 'unilateral']

    # Convert DegreePD into lists (remove degree symbol, split multiple values)
    df['DegreePD'] = df['DegreePD'].astype(str).str.replace('°', '', regex=True).str.split(',')

    print("\n### Processing Eye Misalignments with Normal Head Pose ###")
    total_samples = len(df)
    print(f"Total samples to process: {total_samples}")
    print(f"Starting from index: {start_index}")

    # Use iloc to slice the DataFrame from start_index
    for index, row in df.iloc[start_index:].iterrows():
        try:
            print(f"\n{'='*50}")
            print(f"Processing sample {index + 1}/{total_samples}...")

            # Create and initialize pose tensor with EXACT values for normal head pose
            pose = torch.zeros(15, dtype=torch.float).to(device)
            expr = torch.zeros(100, dtype=torch.float).to(device)

            # Set ONLY pose[1] to np.pi for front view, everything else at 0
            pose[1] = np.pi   # ONLY set this value for normal head

            # Extract and clean values from CSV
            eye_misalignment_list = str(row['EyeMisalignment']).strip().lower().split(',')

            # Ensure eye value is correctly formatted and valid
            eye = str(row['Eye']).strip().lower().replace('\xa0', '')  # Remove hidden spaces
            valid_eyes = ['left', 'right', 'bilateral']

            # Force consistent string comparison
            if eye not in valid_eyes:
                print(f"Final Warning: Unexpected eye value '{eye}' in sample {index + 1}. Skipping...")
                continue  # Skip invalid rows

            degree_pd_list = row['DegreePD']  # Already converted to list

            # Apply Multiple Eye Misalignments
            if len(eye_misalignment_list) == len(degree_pd_list):  # Ensure matching lengths
                for i, eye_misalignment in enumerate(eye_misalignment_list):
                    eye_misalignment = eye_misalignment.strip()  # Ensure no spaces
                    degree_pd = float(degree_pd_list[i].strip())  # Convert to float

                    # Direct manipulation of eye parameters ONLY
                    if eye_misalignment == 'exotropia':
                        # Manually apply exotropia
                        degree_rad = np.radians(degree_pd)
                        if eye == 'left':
                            pose[13] = -degree_rad  # Left eye outward (negative)
                        elif eye == 'right':
                            pose[10] = degree_rad   # Right eye outward (positive)
                        elif eye == 'bilateral':
                            pose[13] = -degree_rad  # Left eye outward
                            pose[10] = degree_rad   # Right eye outward

                    elif eye_misalignment == 'esotropia':
                        # Manually apply esotropia
                        degree_rad = np.radians(degree_pd)
                        if eye == 'left':
                            pose[13] = degree_rad   # Left eye inward (positive)
                        elif eye == 'right':
                            pose[10] = -degree_rad  # Right eye inward (negative)
                        elif eye == 'bilateral':
                            pose[13] = degree_rad   # Left eye inward
                            pose[10] = -degree_rad  # Right eye inward

                    elif eye_misalignment == 'hypertropia':
                        # Manually apply hypertropia
                        degree_rad = np.radians(degree_pd)
                        if eye == 'left':
                            pose[12] = -degree_rad  # Left eye up (negative)
                        elif eye == 'right':
                            pose[9] = -degree_rad   # Right eye up (negative)
                        elif eye == 'bilateral':
                            pose[12] = -degree_rad  # Left eye up
                            pose[9] = -degree_rad   # Right eye up

                    elif eye_misalignment == 'hypotropia':
                        # Manually apply hypotropia
                        degree_rad = np.radians(degree_pd)
                        if eye == 'left':
                            pose[12] = degree_rad   # Left eye down (positive)
                        elif eye == 'right':
                            pose[9] = degree_rad    # Right eye down (positive)
                        elif eye == 'bilateral':
                            pose[12] = degree_rad   # Left eye down
                            pose[9] = degree_rad    # Right eye down

                    else:
                        print(f"Warning: Unrecognized eye misalignment type '{eye_misalignment}'")
            else:
                print(f"Warning: Mismatch between EyeMisalignment and DegreePD for sample {index + 1}. Skipping...")
                continue

            # Double-check that ONLY pose[1] is set and everything else is 0 for head
            for i in range(7):
                if i != 1:  # Skip pose[1] which should be np.pi
                    pose[i] = 0  # Make sure all other head params are 0

            # Generate a single view with fixed head pose
            rgba, shaded_mesh = synthesize_novel_poses_and_expressions(
                expr=expr,
                pose=pose,
                image_size=(512, 512)
            )

            # Create a simple list with just one view for compatibility with save_results
            views_results = [("Front", rgba)]

            # Get case ID from the row
            case_id = row['CaseID']
            diagnose = row['Diagnose']

            # Create directory structure with Diagnose + case ID
            output_dir = os.path.join(base_directory, diagnose, case_id + "_normal_head")
            os.makedirs(output_dir, exist_ok=True)
            # Save metadata in a more organized table-like format
            # Create a complete table-like text representation of metadata
            metadata_table = f"""
            | **Field**          | **Value**                 | **Imputed** |
            |--------------------|---------------------------|-------------|
            | Diagnosis          | {diagnose}                | {"Yes" if row.get('PatientDiagnose_Imputed', '') == 'T' else "No"} |
            | Eye Side           | {row['Eye']}              | {"Yes" if row.get('Eye_Imputed', '') == 'T' else "No"} |
            | Eye Misalignment   | {row['EyeMisalignment']}  | {"Yes" if row.get('EyeMisalignment_Imputed', '') == 'T' else "No"} |
            | Deviation Degree   | {row['DegreePD']}         | {"Yes" if row.get('DegreePD_Imputed', '') == 'T' else "No"} |
            | Case ID            | {case_id}                 | - |
            """

            # Save both formats - JSON for machine reading and text table for human reading
            with open(os.path.join(output_dir, 'metadata.txt'), 'w') as f:
                f.write(metadata_table)


            # Save image
            fig, ax = plt.subplots(figsize=(10, 10))
            ax.imshow(rgba[0, :3].cpu().permute(1, 2, 0))
            ax.axis("off")
            plt.savefig(os.path.join(output_dir, 'front_view.png'), bbox_inches='tight', pad_inches=0)
            plt.close()

            print(f"Completed sample {index + 1}/{total_samples}")

        except Exception as e:
            print(f"Error processing sample {index + 1}: {str(e)}")
            continue
```

```python
##patient view

def process_head_pose_only(data_file, device, base_directory, start_index=0):
    """
    Process head pose adjustments with compensatory eye movements to maintain
    forward gaze direction regardless of head position and camera view.

    Applies eye compensation for:
    - Face/head turn (horizontal)
    - Chin up/down (vertical)

    No eye compensation is applied for head tilt.

    Args:
        data_file: Path to the CSV file
        device: PyTorch device to use
        base_directory: Base directory for saving outputs
        start_index: Index to start processing from (default: 0)
    """
    # Load and process data
    df = pd.read_csv(data_file)

    # Convert AHPDegree into lists (remove degree symbol, split multiple values)
    df['AHPDegree'] = df['AHPDegree'].astype(str).str.replace('°', '', regex=True).str.split(',')

    print("\n### Processing Head Pose with Straight Gaze Across Camera Views ###")
    total_samples = len(df)
    print(f"Total samples to process: {total_samples}")
    print(f"Starting from index: {start_index}")

    # Use iloc to slice the DataFrame from start_index
    for index, row in df.iloc[start_index:].iterrows():
        try:
            print(f"\n{'='*50}")
            print(f"Processing sample {index + 1}/{total_samples}...")

            # Create and initialize pose tensor
            pose = torch.zeros(15, dtype=torch.float).to(device)
            expr = torch.zeros(100, dtype=torch.float).to(device)
            pose[1] = np.pi  # Set front view

            # Extract and clean values from CSV
            ahp_type_list = str(row['AHPType']).strip().lower().split(',')  # Convert to list
            ahp_direction_list = str(row['AHPDirection']).strip().lower().split(',')
            ahp_degree_list = row['AHPDegree']  # Already converted to list

            # Initialize head pose variables
            head_turn_angle = 0.0  # Horizontal rotation (yaw)
            head_tilt_angle = 0.0  # Tilt left/right (roll)
            chin_updown_angle = 0.0  # Chin up/down (pitch)

            # First pass: Track head movement angles
            if len(ahp_type_list) == len(ahp_direction_list) == len(ahp_degree_list):
                for i, ahp_type in enumerate(ahp_type_list):
                    ahp_type = ahp_type.strip()
                    ahp_direction = ahp_direction_list[i].strip()
                    ahp_degree = float(ahp_degree_list[i].strip())

                    # Track head movement angles (in degrees)
                    if ahp_type == 'head turn' or ahp_type == "face turn":
                        if ahp_direction == 'right':
                            head_turn_angle += ahp_degree
                        elif ahp_direction == 'left':
                            head_turn_angle -= ahp_degree

                    elif ahp_type == 'head tilt':
                        if ahp_direction == 'right':
                            head_tilt_angle += ahp_degree
                        elif ahp_direction == 'left':
                            head_tilt_angle -= ahp_degree

                    elif ahp_type == 'chin-up' and ahp_direction == 'upward':
                        chin_updown_angle -= ahp_degree  # Negative for up

                    elif ahp_type == 'chin-down' and ahp_direction == 'downward':
                        chin_updown_angle += ahp_degree  # Positive for down

                # Apply all head movements to the base pose
                if head_turn_angle != 0:
                    direction = 'right' if head_turn_angle > 0 else 'left'
                    pose = set_head_turn(pose, abs(head_turn_angle), direction)

                if head_tilt_angle != 0:
                    direction = 'right' if head_tilt_angle > 0 else 'left'
                    pose = set_head_tilt(pose, abs(head_tilt_angle), direction)


                if chin_updown_angle < 0:  # Chin up
                    pose, expr = set_chin_up(pose, expr, abs(chin_updown_angle))
                elif chin_updown_angle > 0:  # Chin down
                    pose, expr = set_chin_down(pose, expr, abs(chin_updown_angle))

                # Apply fixed eye compensation based on the testing we did
                # This will be consistent across all camera views
                eye_compensation = 0.5  # Using the value that worked best in our tests

                # Apply eye compensation for face turn
                if head_turn_angle > 0:  # Right face turn
                    # Right eye inward, left eye outward
                    pose[10] = eye_compensation
                    pose[13] = eye_compensation
                elif head_turn_angle < 0:  # Left face turn
                    # Left eye inward, right eye outward
                    pose[13] = -eye_compensation
                    pose[10] = -eye_compensation

            else:
                print(f"Warning: Mismatch in AHPType, AHPDirection, and AHPDegree for sample {index + 1}. Skipping...")
                continue

            # Get case ID and diagnosis from the row
            case_id = row['CaseID']
            diagnose = row['Diagnose']

            # Create directory structure with Diagnose + case ID
            output_dir = os.path.join(base_directory, diagnose, case_id + "_straight_gaze")
            os.makedirs(output_dir, exist_ok=True)

            # Create metadata with head angles and eye compensation
            metadata_table = f"""
| **Field**                | **Value**                 | **Imputed** |
|--------------------------|---------------------------|-------------|
| Diagnosis                | {diagnose}                | {"Yes" if row.get('PatientDiagnose_Imputed', '') == 'T' else "No"} |
| AHP Type                 | {row['AHPType']}          | {"Yes" if row.get('AHPType_Imputed', '') == 'T' else "No"} |
| AHP Direction            | {row['AHPDirection']}     | {"Yes" if row.get('AHPDirection_Imputed', '') == 'T' else "No"} |
| AHP Degree               | {row['AHPDegree']}        | {"Yes" if row.get('AHPDegree_Imputed', '') == 'T' else "No"} |
| Case ID                  | {case_id}                 | - |
"""

            # Save metadata
            with open(os.path.join(output_dir, 'metadata.txt'), 'w') as f:
                f.write(metadata_table)

            # Define camera views
            views = [
                ("Frontal", 0, 1, 0),
                ("Up", -0.5, 1, 0),      # Camera lower, looking up
                ("Down", 0.5, 1, 0),     # Camera higher, looking down
                ("Left", 0, 1, 0.75),    # Camera to the left
                ("Middle_Left", 0, 1, 0.4),  # Slightly to the left
                ("Right", 0, 1, -0.75),   # Camera to the right
                ("Middle_Right", 0, 1, -0.4)  # Slightly to the right
            ]

            # Generate views with head pose and compensatory eye movements
            views_results = []
            for view_name, camera_height, camera_distance, camera_side in views:
                # Calculate camera angles
                pitch = np.arctan2(camera_height, camera_distance)
                yaw = np.arctan2(camera_side, camera_distance)

                # Clone and adjust pose for this camera view
                adjusted_pose = pose.clone()
                adjusted_pose[2] += pitch  # Adjust pitch
                adjusted_pose[1] += yaw    # Adjust yaw

                # Perform synthesis
                result = synthesize_novel_poses_and_expressions(
                    expr=expr,
                    pose=adjusted_pose,
                    image_size=(512, 512)
                )

                # Handle tuple return
                if isinstance(result, tuple):
                    result = result[0]

                views_results.append((view_name, result))

                # Save individual view
                fig, ax = plt.subplots(figsize=(10, 10))
                ax.imshow(result[0, :3].cpu().permute(1, 2, 0))
                ax.axis("off")
                view_filename = f"{view_name}.png"
                plt.savefig(os.path.join(output_dir, view_filename), bbox_inches='tight', pad_inches=0)
                plt.close()

            # Save combined view
            fig, axes = plt.subplots(2, 4, figsize=(30, 15))
            axes = axes.flatten()

            for i, (view_name, result) in enumerate(views_results):
                if i < len(axes):
                    axes[i].imshow(result[0, :3].cpu().permute(1, 2, 0))
                    axes[i].axis("off")
                    axes[i].set_title(view_name, fontsize=20)

            if len(views_results) < len(axes):
                for i in range(len(views_results), len(axes)):
                    fig.delaxes(axes[i])

            plt.tight_layout()
            plt.savefig(os.path.join(output_dir, 'combined_views.png'), bbox_inches='tight')
            plt.close()

            print(f"Completed sample {index + 1}/{total_samples} - Saved to {output_dir}")

            # Also save a comparison image showing with and without eye compensation
            # Create a pose without eye compensation
            pose_no_comp = pose.clone()
            pose_no_comp[9] = 0   # Right eye vertical
            pose_no_comp[10] = 0  # Right eye horizontal
            pose_no_comp[12] = 0  # Left eye vertical
            pose_no_comp[13] = 0  # Left eye horizontal

            # Frontal view for comparison
            result_with_comp = synthesize_novel_poses_and_expressions(
                expr=expr,
                pose=pose,
                image_size=(512, 512)
            )
            if isinstance(result_with_comp, tuple):
                result_with_comp = result_with_comp[0]

            result_no_comp = synthesize_novel_poses_and_expressions(
                expr=expr,
                pose=pose_no_comp,
                image_size=(512, 512)
            )
            if isinstance(result_no_comp, tuple):
                result_no_comp = result_no_comp[0]

            # Save comparison
            fig, axes = plt.subplots(ncols=2, figsize=(20, 10))
            axes[0].imshow(result_no_comp[0, :3].cpu().permute(1, 2, 0))
            axes[0].set_title(f"No Eye Compensation", fontsize=16)
            axes[0].axis("off")

            axes[1].imshow(result_with_comp[0, :3].cpu().permute(1, 2, 0))
            axes[1].set_title(f"With Eye Compensation", fontsize=16)
            axes[1].axis("off")

            plt.tight_layout()

            plt.close()

        except Exception as e:
            print(f"Error processing sample {index + 1}: {str(e)}")
            continue
```

```python
import os
import shutil
import traceback

def organize_case_by_case(head_pose_dir, eye_misalignment_dir, target_dir):
    """
    Process each case ID one at a time, copying both head pose and eye misalignment files.

    Args:
        head_pose_dir: Directory containing head pose outputs
        eye_misalignment_dir: Directory containing eye misalignment outputs
        target_dir: Target directory for the organized structure
    """
    # Ensure target directory exists
    os.makedirs(target_dir, exist_ok=True)

    # Check if source directories exist
    if not os.path.exists(head_pose_dir):
        print(f"ERROR: Head pose directory does not exist: {head_pose_dir}")
        return

    if not os.path.exists(eye_misalignment_dir):
        print(f"ERROR: Eye misalignment directory does not exist: {eye_misalignment_dir}")
        return

    # First, collect all case IDs from both directories
    all_cases = set()

    # Get case IDs from head pose directory
    print("Collecting case IDs from head pose directory...")
    head_cases = collect_case_ids(head_pose_dir, "_normal_eyes")
    all_cases.update(head_cases)
    print(f"Found {len(head_cases)} cases in head pose directory")

    # Get case IDs from eye misalignment directory
    print("Collecting case IDs from eye misalignment directory...")
    eye_cases = collect_case_ids(eye_misalignment_dir, "_normal_head")
    all_cases.update(eye_cases)
    print(f"Found {len(eye_cases)} cases in eye misalignment directory")

    print(f"Total unique case IDs: {len(all_cases)}")

    # Process each case one at a time
    processed_count = 0
    for case_id in sorted(all_cases):
        print(f"\nProcessing case: {case_id}")

        # Create target case directory
        target_case_dir = os.path.join(target_dir, f"case_{case_id}")
        os.makedirs(target_case_dir, exist_ok=True)

        # Process head pose files for this case
        head_pose_found = process_case_files(head_pose_dir, target_case_dir, case_id, "_normal_eyes", "head_pose")

        # Process eye misalignment files for this case
        eye_misalignment_found = process_case_files(eye_misalignment_dir, target_case_dir, case_id, "_normal_head", "eye_misalignment")

        # Create case-level metadata
        create_case_metadata(target_case_dir, case_id, head_pose_found, eye_misalignment_found)

        if head_pose_found or eye_misalignment_found:
            processed_count += 1
            print(f"Successfully processed case: {case_id}")
        else:
            print(f"WARNING: No files found for case: {case_id}")

    print(f"\nOrganization complete. Processed {processed_count} cases.")
    print(f"All cases have been organized in {target_dir}")

def collect_case_ids(base_dir, suffix):
    """
    Collect all case IDs from a directory, ensuring consistent format.

    Args:
        base_dir: Base directory to search
        suffix: Suffix to remove from directory names

    Returns:
        Set of case IDs (without 'case_' prefix)
    """
    case_ids = set()

    # Get list of all diagnose directories
    diagnoses = [d for d in os.listdir(base_dir) if os.path.isdir(os.path.join(base_dir, d))]

    for diagnose in diagnoses:
        diagnose_dir = os.path.join(base_dir, diagnose)

        # Check for case directories directly under diagnose directory
        contents = os.listdir(diagnose_dir)
        for item in contents:
            item_path = os.path.join(diagnose_dir, item)
            if os.path.isdir(item_path):
                # Extract the numeric case ID consistently
                case_id = None

                # List of all possible suffixes to check
                all_suffixes = [suffix, "_straight_gaze", "_normal_eyes", "_normal_head"]

                # Remove any suffixes if present
                base_name = item
                for s in all_suffixes:
                    if s and s in base_name:
                        base_name = base_name.replace(s, "")

                # Handle "case_" prefix
                if base_name.startswith("case_"):
                    case_id = base_name[5:]  # Remove "case_" prefix
                else:
                    case_id = base_name

                # Verify it's a valid case ID (not empty after removing prefixes/suffixes)
                if case_id and (case_id.isdigit() or case_id.isalnum()):
                    case_ids.add(case_id)

    return case_ids

def process_case_files(source_base_dir, target_case_dir, case_id, source_suffix, target_subfolder):
    """
    Process files for a specific case.

    Args:
        source_base_dir: Base source directory
        target_case_dir: Target directory for this case
        case_id: Case ID to process (without 'case_' prefix)
        source_suffix: Suffix to look for in source directories
        target_subfolder: Subfolder name for this type of data

    Returns:
        Boolean indicating if files were found and processed
    """
    # Create target subfolder
    target_output_dir = os.path.join(target_case_dir, target_subfolder)
    os.makedirs(target_output_dir, exist_ok=True)

    # Search for the case directory in the source directory
    found = False
    diagnoses = [d for d in os.listdir(source_base_dir) if os.path.isdir(os.path.join(source_base_dir, d))]

    for diagnose in diagnoses:
        diagnose_dir = os.path.join(source_base_dir, diagnose)
        contents = os.listdir(diagnose_dir)

        for item in contents:
            # Define all possible directory naming patterns
            possible_names = [
                f"{case_id}{source_suffix}",           # Like "0495_normal_eyes"
                f"case_{case_id}{source_suffix}",      # Like "case_0495_normal_eyes"
                f"{case_id}_straight_gaze",            # Your head pose naming pattern
                f"case_{case_id}_straight_gaze",       # In case case_ prefix is used
                f"{case_id}_normal_eyes",              # Standard eye pattern
                f"{case_id}_normal_head",              # Standard head pattern
                f"case_{case_id}"                      # Plain case dir
            ]

            if any(pattern == item for pattern in possible_names):
                source_dir = os.path.join(diagnose_dir, item)

                if os.path.isdir(source_dir):
                    # Found the directory, now copy all files
                    files_copied = 0
                    for file in os.listdir(source_dir):
                        source_file = os.path.join(source_dir, file)
                        target_file = os.path.join(target_output_dir, file)

                        if os.path.isfile(source_file):
                            shutil.copy2(source_file, target_file)
                            files_copied += 1

                    if files_copied > 0:
                        print(f"  Copied {files_copied} files for {target_subfolder}")
                        print(f"  Source directory: {source_dir}")
                        found = True
                        break

        if found:
            break

    if not found:
        print(f"  No {target_subfolder} files found for case {case_id}")

    return found

def create_case_metadata(target_case_dir, case_id, head_pose_found, eye_misalignment_found):
    """
    Create metadata file for a case.

    Args:
        target_case_dir: Target directory for this case
        case_id: Case ID
        head_pose_found: Whether head pose files were found
        eye_misalignment_found: Whether eye misalignment files were found
    """
    # Try to determine diagnosis from available metadata files
    diagnose_value = "Unknown"

    # Check in head pose directory first
    if head_pose_found:
        head_pose_dir = os.path.join(target_case_dir, "head_pose")
        metadata_file = os.path.join(head_pose_dir, "metadata.txt")

        if os.path.exists(metadata_file):
            diagnose_value = extract_diagnosis_from_metadata(metadata_file)

    # If still unknown, check eye misalignment directory
    if diagnose_value == "Unknown" and eye_misalignment_found:
        eye_dir = os.path.join(target_case_dir, "eye_misalignment")
        metadata_file = os.path.join(eye_dir, "metadata.txt")

        if os.path.exists(metadata_file):
            diagnose_value = extract_diagnosis_from_metadata(metadata_file)

    # Create case metadata
    case_metadata = f"""
| **Field**          | **Value**                 |
|--------------------|---------------------------|
| Case ID            | case_{case_id}            |
| Diagnosis          | {diagnose_value}          |
| Head Pose Files    | {"Yes" if head_pose_found else "No"} |
| Eye Misalignment Files | {"Yes" if eye_misalignment_found else "No"} |
"""

    with open(os.path.join(target_case_dir, "case_metadata.txt"), "w") as f:
        f.write(case_metadata)

def extract_diagnosis_from_metadata(metadata_file):
    """
    Extract diagnosis from metadata file.

    Args:
        metadata_file: Path to metadata file

    Returns:
        Diagnosis value if found, "Unknown" otherwise
    """
    try:
        with open(metadata_file, "r") as f:
            content = f.read()

            if "Diagnosis" in content:
                for line in content.split("\n"):
                    if "Diagnosis" in line and "|" in line:
                        parts = line.split("|")
                        if len(parts) >= 3:
                            return parts[2].strip()
    except Exception:
        pass

    return "Unknown"
```
